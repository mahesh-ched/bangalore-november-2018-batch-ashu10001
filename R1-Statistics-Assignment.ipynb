# Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (7 points)

Goal_diff_count = GoalsFor - GoalsAgainst

Assumption: Considering the absolute value of the difference 

```python
#Function to find the goal difference
def Goal_diff_count(goals_f,goals_a):
    Goal_diff_cnt = abs(goals_f - goals_a)
    return (Goal_diff_cnt)

```

```python
#Adding new column to df2
df2["Goal_difference"] = Goal_diff_count(pd.to_numeric(df2["GoalsFor"]),pd.to_numeric(df2["GoalsAgainst"]))
```

```python
#All the teams with goal differences
df2.loc[:,["Team","Goal_difference"]]
```

```python
#Index of the item with max goal difference
max_idx = Goal_diff_count(pd.to_numeric(df2["GoalsFor"]),pd.to_numeric(df2["GoalsAgainst"])).idxmax()

#Team name of the team with max goal difference
df2.iloc[max_idx-1,1]
```

```python
#Goal_diff_count(pd.to_numeric(df2["GoalsFor"]),pd.to_numeric(df2["GoalsAgainst"])).idxmax()
#Index of the item with least goal difference
min_idx = Goal_diff_count(pd.to_numeric(df2["GoalsFor"]),pd.to_numeric(df2["GoalsAgainst"])).idxmin()

#Team name of the team with least goal difference
df2.iloc[min_idx-1,1]
```

# Create a new column with name “Winning Percent” and append it to the data set 

Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

```python
df2["Winning Percent"] = (pd.to_numeric(df2["GamesWon"])/pd.to_numeric(df2["GamesPlayed"]))*100
```

# Group teams based on their “Best position” and print the sum of their points for all positions 

Eg: Best Position                Points

        1                              25000

        2                              7000  

```python
#Changing the datatype of columns 'Points' and 'BestPosition' to int
df2["Points"] = df2.Points.astype(int)
df2["BestPosition"] = df2.BestPosition.astype(int)
```

```python
df_agg = df2.groupby("BestPosition")["Points"].agg({"Points":"sum"})
```

```python
#resetting the index 
df_agg = df_agg.reset_index()
```

```python
#sorting by BestPosition
df_agg.sort_values(by="BestPosition")
```
